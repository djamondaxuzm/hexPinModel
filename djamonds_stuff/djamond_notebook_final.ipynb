```python
import os
os.environ["HDF5_USE_FILE_LOCKING"] = "FALSE"

import numpy as np
import matplotlib.pyplot as plt
import openmc

# ----------------------------------------------------------------------
# Cross-section library
# ----------------------------------------------------------------------
XS_PATH = "/mnt/c/Users/djamo/Desktop/OpenMC XS/endfb-viii.0-hdf5/cross_sections.xml"
openmc.config['cross_sections'] = XS_PATH

# ----------------------------------------------------------------------
# Global parameters (units: cm, K)
# ----------------------------------------------------------------------
R_FUEL = 0.60     # fuel radius (cm)
R_CLAD = 0.62     # cladding outer radius (cm)
H      = 100.0    # axial height of model domain (cm), reflective
T_REF  = 600.0    # temperature (K) for all materials

# ----------------------------------------------------------------------
# Materials
# ----------------------------------------------------------------------
def make_materials():
    fuel = openmc.Material(name='uo2')
    fuel.add_nuclide('U235', 0.02)
    fuel.add_nuclide('U238', 0.98)
    fuel.add_nuclide('O16', 5.0)
    fuel.set_density('g/cm3', 11.5)   # example UO2 density
    fuel.temperature = T_REF

    cladding = openmc.Material(name='zircaloy4')
    cladding.add_element('Zr', 0.98)
    cladding.add_element('Sn', 0.015)
    cladding.add_element('Fe', 0.002)
    cladding.add_element('Cr', 0.002)
    cladding.add_element('O', 0.001)
    cladding.set_density('g/cm3', 6.55)
    cladding.temperature = T_REF

    water = openmc.Material(name='h2o')
    water.add_nuclide('H1', 2.0)
    water.add_nuclide('O16', 1.0)
    water.set_density('g/cm3', 0.70)  # ~PWR hot density (approx)
    water.add_s_alpha_beta('c_H_in_H2O')
    water.temperature = T_REF

    boron_ppm_const = 2000
    water_material = openmc.model.borated_water(
        boron_ppm = boron_ppm_const, 
        temperature = T_REF, 
        pressure = 15
    )

    mats = openmc.Materials([fuel, cladding, water])
    mats.cross_sections = XS_PATH
    return mats, fuel, cladding, water

# ----------------------------------------------------------------------
# Fuel-to-moderator ratio for a hex cell (2D)
# ----------------------------------------------------------------------
def fuel_to_moderator_ratio(pitch_cm: float) -> float:
    """
    Fuel-to-moderator volume ratio for a single hex cell with face-to-face
    pitch 'pitch_cm':

      A_hex  = (sqrt(3)/2) * pitch^2
      A_fuel = π R_FUEL^2
      A_clad = π (R_CLAD^2 - R_FUEL^2)
      A_mod  = A_hex - A_fuel - A_clad

      V_fuel / V_mod = A_fuel / A_mod
    """
    A_hex  = np.sqrt(3.0) / 2.0 * pitch_cm**2
    A_fuel = np.pi * R_FUEL**2
    A_clad = np.pi * (R_CLAD**2 - R_FUEL**2)
    A_mod  = A_hex - A_fuel - A_clad

    if A_mod <= 0.0:
        raise ValueError(f"Nonpositive moderator area at pitch={pitch_cm} cm")

    return A_fuel / A_mod

# ----------------------------------------------------------------------
# Build OpenMC model for a given hex pitch
# ----------------------------------------------------------------------
def build_model(pitch_cm: float) -> openmc.Model:
    """
    Build an infinite hexagonal pin-cell model with given face-to-face pitch.
    Reflective boundaries on the hex prism and in z.
    """
    mats, fuel, cladding, water = make_materials()

    # Surfaces
    z_min = openmc.ZPlane(z0=-H/2.0, boundary_type='reflective')
    z_max = openmc.ZPlane(z0= H/2.0, boundary_type='reflective')

    cyl_fuel = openmc.ZCylinder(r=R_FUEL)
    cyl_clad = openmc.ZCylinder(r=R_CLAD)

    # Hexagonal prism boundary: edge length = pitch / sqrt(3)
    edge_length = pitch_cm / np.sqrt(3.0)
    hex_region = openmc.model.hexagonal_prism(
        edge_length=edge_length,
        orientation='y',
        boundary_type='reflective'
    )

    # Regions / cells inside the hex cell
    fuel_region  = -cyl_fuel
    clad_region  = +cyl_fuel & -cyl_clad
    water_region = +cyl_clad

    fuel_cell  = openmc.Cell(name='fuel',      region=fuel_region,  fill=fuel)
    clad_cell  = openmc.Cell(name='cladding',  region=clad_region,  fill=cladding)
    water_cell = openmc.Cell(name='moderator', region=water_region, fill=water)

    pin_univ = openmc.Universe(cells=[fuel_cell, clad_cell, water_cell])

    # Root cell: hex prism in r, reflective planes in z, filled with pin universe
    root_region = hex_region & -z_max & +z_min
    root_cell   = openmc.Cell(region=root_region, fill=pin_univ)
    root_univ   = openmc.Universe(cells=[root_cell])

    geom = openmc.Geometry(root_univ)

    # Settings
    settings = openmc.Settings()
    settings.run_mode  = 'eigenvalue'
    settings.batches   = 100
    settings.inactive  = 50
    settings.particles = 10_000

    settings.temperature = {
        'default': T_REF,
        'method': 'interpolation',
        'range': (300.0, 1500.0)
    }

    # Source: keep it safely inside the hex cell (circle of radius R_CLAD)
    r_box = R_CLAD
    lower_left  = (-r_box, -r_box, -H/2.0)
    upper_right = ( r_box,  r_box,  H/2.0)
    source_space = openmc.stats.Box(lower_left, upper_right)
    settings.source = openmc.IndependentSource(space=source_space)

    model = openmc.Model(geometry=geom, materials=mats, settings=settings)
    return model

# ----------------------------------------------------------------------
# Sweep over moderation (hex pitch) and collect k_eff
# ----------------------------------------------------------------------
if __name__ == "__main__":
    # Pitches to sample (cm); must be large enough that moderator area > 0
    pitches = np.linspace(2.2, 2.7, 8)  # adjust as needed

    fm_ratios   = []
    keff_means  = []
    keff_stds   = []

    for pitch in pitches:
        print(f"\n=== Running pitch = {pitch:.3f} cm ===")
        fm_ratio = fuel_to_moderator_ratio(pitch)
        fm_ratios.append(fm_ratio)

        model = build_model(pitch)
        sp_filename = model.run()
        sp = openmc.StatePoint(sp_filename, autolink=False)

        # Extract k_eff
        try:
            k_var  = sp.keff
            k_mean = k_var.nominal_value
            k_std  = k_var.std_dev
        except AttributeError:
            k_mean, k_std = sp.k_combined

        keff_means.append(k_mean)
        keff_stds.append(k_std)

        print(f"  Fuel/moderator ratio = {fm_ratio:.4f}")
        print(f"  k_eff = {k_mean:.5f} ± {k_std:.5f}")
        sp.close()

    fm_ratios  = np.array(fm_ratios)
    keff_means = np.array(keff_means)
    keff_stds  = np.array(keff_stds)

    # ------------------------------------------------------------------
    # Find approximate optimal moderation (max k_eff)
    # ------------------------------------------------------------------
    idx_max    = np.argmax(keff_means)
    best_pitch = pitches[idx_max]
    best_ratio = fm_ratios[idx_max]
    best_keff  = keff_means[idx_max]

    print("\n=== Approximate optimal moderation ===")
    print(f"  pitch*       ≈ {best_pitch:.3f} cm")
    print(f"  (fuel/mod)*  ≈ {best_ratio:.4f}")
    print(f"  k_eff,max    ≈ {best_keff:.5f}")

    # ------------------------------------------------------------------
    # Plot k_eff vs fuel-to-moderator ratio
    # ------------------------------------------------------------------
    plt.errorbar(pitches, keff_means, yerr=keff_stds, fmt="o-")
    plt.xlabel("Pitch (cm)")
    plt.ylabel(r"$k_{\mathrm{eff}}$")
    plt.title("k_eff vs Pitch (hex pin cell, reflective)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import os
os.environ["HDF5_USE_FILE_LOCKING"] = "FALSE"
import numpy as np
import matplotlib.pyplot as plt
import openmc

# ----------------------------------------------------------------------
# Cross section library
# ----------------------------------------------------------------------
XS_PATH = "/mnt/c/Users/djamo/Desktop/OpenMC XS/endfb-viii.0-hdf5/cross_sections.xml"
openmc.config['cross_sections'] = XS_PATH

# ----------------------------------------------------------------------
# Helper functions: temperature-dependent densities
# ----------------------------------------------------------------------

# Try to use IAPWS-97 for water at 15 MPa; fallback to simple approximation
# (Not used in LOCA geometry, but left here for completeness)
try:
    from iapws import IAPWS97
    HAVE_IAPWS = True
except ImportError:
    HAVE_IAPWS = False

def water_density_15MPa_g_cm3(T_K: float) -> float:
    """
    Liquid water density [g/cm^3] at 15 MPa as a function of temperature [K].
    Uses IAPWS-97 if available; otherwise uses a simple linear fit between 300–350°C.
    In this LOCA model, coolant regions are void; this correlation is not used
    in the geometry, only available if needed for other studies.
    """
    if HAVE_IAPWS:
        w = IAPWS97(P=15.0, T=T_K)      # P in MPa, T in K
        return w.rho / 1000.0           # kg/m^3 -> g/cm^3
    else:
        T1, rho1 = 573.15, 0.72
        T2, rho2 = 623.15, 0.65
        if T_K <= T1:
            return rho1
        elif T_K >= T2:
            return rho2
        else:
            return rho1 + (rho2 - rho1) * (T_K - T1) / (T2 - T1)

def fuel_density_g_cm3(T_K: float,
                       T_ref: float = 600.0,
                       rho_ref: float = 11.5) -> float:
    """
    Simple fuel volumetric thermal expansion model for UO2(+Gd):
    rho(T) = rho_ref / (1 + alpha_vol * (T - T_ref)).
    """
    alpha_lin = 1.0e-5           # ~1e-5 1/K linear coefficient
    alpha_vol = 3.0 * alpha_lin  # volumetric
    return rho_ref / (1.0 + alpha_vol * (T_K - T_ref))

# ----------------------------------------------------------------------
# Reference temperature for initial material definition
# ----------------------------------------------------------------------
T_ref = 600.0  # K

# ----------------------------------------------------------------------
# Materials
# ----------------------------------------------------------------------
fuel = openmc.Material(name='uo2_gad')
fuel.add_nuclide('U235', 0.04)
fuel.add_nuclide('U238', 0.96)
fuel.add_nuclide('O16', 2.0)
fuel.set_density('g/cm3', fuel_density_g_cm3(T_ref))
fuel.temperature = T_ref

cladding = openmc.Material(name='zircaloy4')
cladding.add_element('Zr', 0.98)
cladding.add_element('Sn', 0.015)
cladding.add_element('Fe', 0.002)
cladding.add_element('Cr', 0.001)
cladding.add_element('O', 0.001)
cladding.add_element('Hf', 0.001)
cladding.set_density('g/cm3', 6.34)
cladding.temperature = T_ref

# Water material is defined but NOT used in the LOCA geometry (coolant void).
water = openmc.Material(name='h2o')
water.add_nuclide('H1', 2.0)
water.add_nuclide('O16', 1.0)
water.set_density('g/cm3', water_density_15MPa_g_cm3(T_ref))
water.add_s_alpha_beta('c_H_in_H2O')
water.temperature = T_ref

boron_ppm_const = 1000
water_material = openmc.model.borated_water(
    boron_ppm = boron_ppm_const,
    temperature = T_ref,
    pressure = 15
)
# NOTE: also not used in LOCA geometry.

ctrl_rod = openmc.Material(name='bc4')
ctrl_rod.add_nuclide('B10', 0.7)
ctrl_rod.add_nuclide('B11', 0.3)
ctrl_rod.add_nuclide('C12', 4.0)
ctrl_rod.set_density('g/cm3', 2.5)
ctrl_rod.temperature = T_ref

rpv = openmc.Material(name='stainless_steel')
rpv.add_element('Fe', 0.7)
rpv.add_element('Cr', 0.2)
rpv.add_element('Ni', 0.08)
rpv.add_element('Mn', 0.01)
rpv.add_element('C', 0.01)
rpv.set_density('g/cm3', 7.6)
rpv.temperature = T_ref

mats = openmc.Materials([fuel, cladding, water, ctrl_rod, rpv])
mats.cross_sections = XS_PATH

# ----------------------------------------------------------------------
# Geometry (LOCA: all moderator regions are void)
# ----------------------------------------------------------------------
H_core = 225.0
R_core = 70.0   # cm

cyl_uo2   = openmc.ZCylinder(r=0.6)
cyl_clad  = openmc.ZCylinder(r=0.62)
cyl_ctrl  = openmc.ZCylinder(r=0.62)
cyl_no_ctrl = openmc.ZCylinder(r=0.62)
cyl_rpv_i = openmc.ZCylinder(r=R_core - 10.0)
cyl_rpv_o = openmc.ZCylinder(r=R_core, boundary_type='vacuum')
z_max = openmc.ZPlane(z0=H_core,  boundary_type='vacuum')
z_min = openmc.ZPlane(z0=-H_core, boundary_type='vacuum')

uo2_region           = -cyl_uo2     & -z_max & +z_min
clad_region          = +cyl_uo2     & -cyl_clad & -z_max & +z_min
water_region         = +cyl_clad    & -z_max & +z_min
ctrl_region          = -cyl_ctrl    & -z_max & +z_min
no_ctrl_region       = -cyl_no_ctrl & -z_max & +z_min
water_ctrl_region    = +cyl_ctrl    & -z_max & +z_min
water_no_ctrl_region = +cyl_no_ctrl & -z_max & +z_min
rpv_region           = +cyl_rpv_i   & -cyl_rpv_o & -z_max & +z_min
water_rod_region     = -cyl_ctrl    & -z_max & +z_min

# Fuel, cladding, control rod, and vessel are still filled with materials.
uo2_cell       = openmc.Cell(name='fuel',      region=uo2_region,      fill=fuel)
clad_cell      = openmc.Cell(name='cladding',  region=clad_region,     fill=cladding)
ctrl_cell      = openmc.Cell(name='ctrl_rod',  region=ctrl_region,     fill=ctrl_rod)
rpv_cell       = openmc.Cell(name='vessel',    region=rpv_region,      fill=rpv)

# LOCA: coolant / moderator regions are void (no fill).
no_ctrl_cell       = openmc.Cell(name='no_ctrl_rod', region=no_ctrl_region)  # void
water_cell         = openmc.Cell(name='void_coolant', region=water_region)   # void
water_cell_ctrl    = openmc.Cell(name='void_ctrl', region=water_ctrl_region)
water_cell_no_ctrl = openmc.Cell(name='void_no_ctrl', region=water_no_ctrl_region)
water_rod_cell     = openmc.Cell(name='void_rod_region', region=water_rod_region)

fuel_universe        = openmc.Universe(cells=[uo2_cell, clad_cell, water_cell])
ctrl_rod_universe    = openmc.Universe(cells=[ctrl_cell, water_cell_ctrl])
no_ctrl_rod_universe = openmc.Universe(cells=[no_ctrl_cell, water_cell_no_ctrl])

# Outer universe is also void (no moderator downcomer in this LOCA approximation)
outer_universe = openmc.Universe(cells=[openmc.Cell()])

lat = openmc.HexLattice()
lat.center = (0.0, 0.0)
lat.pitch  = (2.35,)
lat.outer  = outer_universe

outer_ring  = [fuel_universe] * 12
middle_ring = [fuel_universe] * 6
inner_ring  = [ctrl_rod_universe]
lat.universes = [outer_ring, middle_ring, inner_ring]

a = 2.75 * lat.pitch[0]
outer_boundary = openmc.model.hexagonal_prism(edge_length=a, orientation='y')
main_cell = openmc.Cell(fill=lat, region=outer_boundary & -z_max & +z_min)
assembly_univ = openmc.Universe(cells=[main_cell])

core_lat = openmc.HexLattice()
core_lat.center = (0.0, 0.0)
core_lat.pitch  = (np.sqrt(3.0) * a,)
core_lat.outer  = outer_universe
core_lat.orientation = 'x'

ring_1 = [assembly_univ]
ring_2 = [assembly_univ] * 6
ring_3 = [assembly_univ] * 12
ring_4 = [assembly_univ] * 18
ring_5 = [assembly_univ] * 24
ring_6 = [assembly_univ] * 30
core_lat.universes = [ring_6, ring_5, ring_4, ring_3, ring_2, ring_1]

whole_core_cell = openmc.Cell(fill=core_lat,
                              region=-cyl_rpv_i & -z_max & +z_min)
geom = openmc.Geometry([whole_core_cell, rpv_cell])

# ----------------------------------------------------------------------
# Settings
# ----------------------------------------------------------------------
settings = openmc.Settings()
settings.batches   = 100
settings.inactive  = 50
settings.particles = 100_000
settings.run_mode  = 'eigenvalue'

settings.temperature = {
    'method': 'interpolation',
    'default': T_ref,
    'range': (300.0, 1500.0)
}

# Source distribution: box matching the core region
core_radius = R_core - 10.0      # same as inner vessel radius (cyl_rpv_i)
lower_left  = (-core_radius, -core_radius, -H_core)
upper_right = ( core_radius,  core_radius,  H_core)

source_space = openmc.stats.Box(lower_left, upper_right)
settings.source = openmc.IndependentSource(space=source_space)

# ----------------------------------------------------------------------
# Model
# ----------------------------------------------------------------------
model = openmc.Model(geometry=geom, materials=mats, settings=settings)

# ----------------------------------------------------------------------
# Report basic core geometry
# ----------------------------------------------------------------------
core_area_cm2 = np.pi * R_core**2           # cm^2
core_area_m2  = core_area_cm2 * 1.0e-4      # m^2

print("LOCA core geometric parameters:")
print(f"  Core radius            = {R_core:.2f} cm")
print(f"  Core cross-sectional A = {core_area_cm2:.2f} cm^2 "
      f"({core_area_m2:.4f} m^2)")
print("  Coolant / moderator regions are void in this model (LOCA).\n")

# ----------------------------------------------------------------------
# Helper: isothermal state (all materials at same T, coolant void)
# ----------------------------------------------------------------------
def set_thermal_state_isothermal(T_iso: float) -> None:
    """
    Set an isothermal state for LOCA case:
      - Fuel, cladding, control, and RPV all at T_iso.
      - Coolant regions are void (no material), so water is not used.
    """
    fuel.temperature     = T_iso
    cladding.temperature = T_iso
    ctrl_rod.temperature = T_iso
    rpv.temperature      = T_iso

    # Fuel density changes with temperature
    fuel.set_density('g/cm3', fuel_density_g_cm3(T_iso))

    # Default temperature for interpolation
    settings.temperature['default'] = T_iso

# ----------------------------------------------------------------------
# Isothermal temperature sweep: T_fuel = T_clad = T_ctrl = T_RPV
# (coolant void)
# ----------------------------------------------------------------------
temps_iso = np.linspace(373.0, 773.0, 5)  # K (example: 100°C–500°C)

keff_means_iso = []
keff_stds_iso  = []

print("Isothermal LOCA sweep (all solid materials at same T, coolant void):")
for T in temps_iso:
    set_thermal_state_isothermal(T)

    sp_filename = model.run()
    sp = openmc.StatePoint(sp_filename, autolink=False)

    try:
        k_var  = sp.keff        # newer API: uncertainties.UFloat
        k_mean = k_var.nominal_value
        k_std  = k_var.std_dev
    except AttributeError:
        k_mean, k_std = sp.k_combined  # older API

    sp.close()

    keff_means_iso.append(k_mean)
    keff_stds_iso.append(k_std)

    print(f"  T = {T:7.2f} K : k_eff = {k_mean:.5f} ± {k_std:.5f}")

temps_iso      = np.array(temps_iso)
keff_means_iso = np.array(keff_means_iso)
keff_stds_iso  = np.array(keff_stds_iso)

# ----------------------------------------------------------------------
# Plot k_eff vs temperature (isothermal LOCA)
# ----------------------------------------------------------------------
plt.errorbar(temps_iso, keff_means_iso, yerr=keff_stds_iso, fmt="o-")
plt.xlabel("Isothermal temperature T [K]")
plt.ylabel(r"$k_{\mathrm{eff}}$ [-]")
plt.title("LOCA case: $k_{eff}$ vs. isothermal temperature (coolant void)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
os.environ["HDF5_USE_FILE_LOCKING"] = "FALSE"
import numpy as np
import matplotlib.pyplot as plt
import openmc

# ----------------------------------------------------------------------
# Cross section library
# ----------------------------------------------------------------------
XS_PATH = "/mnt/c/Users/djamo/Desktop/OpenMC XS/endfb-viii.0-hdf5/cross_sections.xml"
openmc.config['cross_sections'] = XS_PATH

# ----------------------------------------------------------------------
# Helper functions: temperature-dependent densities
# ----------------------------------------------------------------------

# Try to use IAPWS-97 for water at 15 MPa; fallback to simple approximation
try:
    from iapws import IAPWS97
    HAVE_IAPWS = True
except ImportError:
    HAVE_IAPWS = False

def water_density_15MPa_g_cm3(T_K: float) -> float:
    """
    Liquid water density [g/cm^3] at 15 MPa as a function of temperature [K].
    Uses IAPWS-97 if available; otherwise uses a simple linear fit between 300–350°C.
    """
    if HAVE_IAPWS:
        w = IAPWS97(P=15.0, T=T_K)      # P in MPa, T in K
        return w.rho / 1000.0           # kg/m^3 -> g/cm^3
    else:
        # Approximate near 15 MPa:
        # ~0.72 g/cm3 at 300°C (573 K); ~0.65 g/cm3 at 350°C (623 K)
        T1, rho1 = 573.15, 0.72
        T2, rho2 = 623.15, 0.65
        if T_K <= T1:
            return rho1
        elif T_K >= T2:
            return rho2
        else:
            return rho1 + (rho2 - rho1) * (T_K - T1) / (T2 - T1)

def fuel_density_g_cm3(T_K: float,
                       T_ref: float = 600.0,
                       rho_ref: float = 11.5) -> float:
    """
    Simple fuel volumetric thermal expansion model for UO2(+Gd):
    rho(T) = rho_ref / (1 + alpha_vol * (T - T_ref)).
    """
    alpha_lin = 1.0e-5           # ~1e-5 1/K linear coefficient
    alpha_vol = 3.0 * alpha_lin  # volumetric
    return rho_ref / (1.0 + alpha_vol * (T_K - T_ref))

# ----------------------------------------------------------------------
# Reference temperature for initial material definition
# ----------------------------------------------------------------------
T_ref = 600.0  # K

# ----------------------------------------------------------------------
# Materials
# ----------------------------------------------------------------------
fuel = openmc.Material(name='uo2_gad')
fuel.add_nuclide('U235', 0.04)
fuel.add_nuclide('U238', 0.96)
fuel.add_nuclide('O16', 2.0)
fuel.set_density('g/cm3', fuel_density_g_cm3(T_ref))
fuel.temperature = T_ref

cladding = openmc.Material(name='zircaloy4')
cladding.add_element('Zr', 0.98)
cladding.add_element('Sn', 0.015)
cladding.add_element('Fe', 0.002)
cladding.add_element('Cr', 0.001)
cladding.add_element('O', 0.001)
cladding.add_element('Hf', 0.001)
cladding.set_density('g/cm3', 6.34)
cladding.temperature = T_ref

water = openmc.Material(name='h2o')
water.add_nuclide('H1', 2.0)
water.add_nuclide('O16', 1.0)
water.set_density('g/cm3', water_density_15MPa_g_cm3(T_ref))
water.add_s_alpha_beta('c_H_in_H2O')
water.temperature = T_ref

boron_ppm_const = 1000
water_material = openmc.model.borated_water(
    boron_ppm = boron_ppm_const,
    temperature = T_ref,
    pressure = 15
)
# NOTE: currently not used in geometry; moderator is 'water' above.

ctrl_rod = openmc.Material(name='b4c')
ctrl_rod.add_nuclide('B10', 2.8)
ctrl_rod.add_nuclide('B11', 1.2)
ctrl_rod.add_nuclide('C12', 1.0)
ctrl_rod.set_density('g/cm3', 2.5)
ctrl_rod.temperature = T_ref

rpv = openmc.Material(name='stainless_steel')
rpv.add_element('Fe', 0.7)
rpv.add_element('Cr', 0.2)
rpv.add_element('Ni', 0.08)
rpv.add_element('Mn', 0.01)
rpv.add_element('C', 0.01)
rpv.set_density('g/cm3', 7.6)
rpv.temperature = T_ref

mats = openmc.Materials([fuel, cladding, water, ctrl_rod, rpv])
mats.cross_sections = XS_PATH

# ----------------------------------------------------------------------
# Geometry
# ----------------------------------------------------------------------
H_core = 225.0
R_core = 70.0   # cm

cyl_uo2   = openmc.ZCylinder(r=0.6)
cyl_clad  = openmc.ZCylinder(r=0.62)
cyl_ctrl  = openmc.ZCylinder(r=0.62)
cyl_no_ctrl = openmc.ZCylinder(r=0.62)
cyl_rpv_i = openmc.ZCylinder(r=R_core - 10.0)
cyl_rpv_o = openmc.ZCylinder(r=R_core, boundary_type='vacuum')
z_max = openmc.ZPlane(z0=H_core,  boundary_type='vacuum')
z_min = openmc.ZPlane(z0=-H_core, boundary_type='vacuum')

uo2_region           = -cyl_uo2     & -z_max & +z_min
clad_region          = +cyl_uo2     & -cyl_clad & -z_max & +z_min
water_region         = +cyl_clad    & -z_max & +z_min
ctrl_region          = -cyl_ctrl    & -z_max & +z_min
no_ctrl_region       = -cyl_no_ctrl & -z_max & +z_min
water_ctrl_region    = +cyl_ctrl    & -z_max & +z_min
water_no_ctrl_region = +cyl_no_ctrl & -z_max & +z_min
rpv_region           = +cyl_rpv_i   & -cyl_rpv_o & -z_max & +z_min
water_rod_region     = -cyl_ctrl    & -z_max & +z_min

uo2_cell          = openmc.Cell(name='fuel',      region=uo2_region,      fill=fuel)
clad_cell         = openmc.Cell(name='cladding',  region=clad_region,     fill=cladding)
ctrl_cell         = openmc.Cell(name='ctrl_rod',  region=ctrl_region,     fill=ctrl_rod)
no_ctrl_cell      = openmc.Cell(name='no_ctrl_rod', region=no_ctrl_region, fill=water)
water_cell        = openmc.Cell(name='water',     region=water_region,    fill=water)
water_cell_ctrl   = openmc.Cell(name='water_ctrl',
                                region=water_ctrl_region, fill=water)
water_cell_no_ctrl = openmc.Cell(name='water_no_ctrl',
                                 region=water_no_ctrl_region, fill=water)
rpv_cell          = openmc.Cell(name='vessel',    region=rpv_region,      fill=rpv)
water_rod_cell    = openmc.Cell(name='water_rod_uncontrolled',
                                region=water_rod_region, fill=water)

fuel_universe         = openmc.Universe(cells=[uo2_cell, clad_cell, water_cell])
ctrl_rod_universe     = openmc.Universe(cells=[ctrl_cell, water_cell_ctrl])
outer_universe        = openmc.Universe(cells=[openmc.Cell(fill=water)])
no_ctrl_rod_universe  = openmc.Universe(cells=[no_ctrl_cell, water_cell_no_ctrl])

lat = openmc.HexLattice()
lat.center = (0.0, 0.0)
lat.pitch  = (2.35,)
lat.outer  = outer_universe

outer_ring  = [fuel_universe] * 12
middle_ring = [fuel_universe] * 6
inner_ring  = [no_ctrl_rod_universe]
lat.universes = [outer_ring, middle_ring, inner_ring]

a = 2.75 * lat.pitch[0]
outer_boundary = openmc.model.hexagonal_prism(edge_length=a, orientation='y')
main_cell = openmc.Cell(fill=lat, region=outer_boundary & -z_max & +z_min)
assembly_univ = openmc.Universe(cells=[main_cell])

core_lat = openmc.HexLattice()
core_lat.center = (0.0, 0.0)
core_lat.pitch  = (np.sqrt(3.0) * a,)
core_lat.outer  = outer_universe
core_lat.orientation = 'x'

ring_1 = [assembly_univ]
ring_2 = [assembly_univ] * 6
ring_3 = [assembly_univ] * 12
ring_4 = [assembly_univ] * 18
ring_5 = [assembly_univ] * 24
ring_6 = [assembly_univ] * 30
core_lat.universes = [ring_6, ring_5, ring_4, ring_3, ring_2, ring_1]

whole_core_cell = openmc.Cell(fill=core_lat,
                              region=-cyl_rpv_i & -z_max & +z_min)
geom = openmc.Geometry([whole_core_cell, rpv_cell])

# ----------------------------------------------------------------------
# Settings
# ----------------------------------------------------------------------
settings = openmc.Settings()
settings.batches   = 100
settings.inactive  = 50
settings.particles = 100_000
settings.run_mode  = 'eigenvalue'

settings.temperature = {
    'method': 'interpolation',
    'default': T_ref,
    'range': (300.0, 1500.0)
}

# Source distribution: box matching the core region
core_radius = R_core - 10.0      # same as inner vessel radius (cyl_rpv_i)
lower_left  = (-core_radius, -core_radius, -H_core)
upper_right = ( core_radius,  core_radius,  H_core)

source_space = openmc.stats.Box(lower_left, upper_right)
settings.source = openmc.IndependentSource(space=source_space)

# ----------------------------------------------------------------------
# Model
# ----------------------------------------------------------------------
model = openmc.Model(geometry=geom, materials=mats, settings=settings)

# ----------------------------------------------------------------------
# Temperature ranges used in alpha calculations
# ----------------------------------------------------------------------
# 1) Isothermal sweep
temps_iso = np.linspace(373.0, 773.0, 2)  # K

# 2) Fuel temperature coefficient (FTC)
T_mod_fixed   = 600.0  # K (moderator / structural reference)
T_struct_FTC  = 600.0  # K
temps_fuel    = np.linspace(293.0, 773.0, 3)  # K

# 3) Moderator temperature coefficient (MTC)
T_fuel_fixed  = 600.0  # K
T_struct_MTC  = 600.0  # K
temps_mod     = np.linspace(473.0, 773.0, 2)  # K

# ----------------------------------------------------------------------
# Print core cross-section and densities at ends of ranges (fuel & mod)
# ----------------------------------------------------------------------
core_area_cm2 = np.pi * R_core**2           # cm^2
core_area_m2  = core_area_cm2 * 1.0e-4      # m^2

print("Core geometric parameters:")
print(f"  Core radius            = {R_core:.2f} cm")
print(f"  Core cross-sectional A = {core_area_cm2:.2f} cm^2 "
      f"({core_area_m2:.4f} m^2)")

# Isothermal range
T_iso_min, T_iso_max = temps_iso[0], temps_iso[-1]
# Fuel sweep range
T_fuel_min, T_fuel_max = temps_fuel[0], temps_fuel[-1]
# Moderator sweep range
T_mod_min, T_mod_max = temps_mod[0], temps_mod[-1]

print("\nDensity summary for alpha sweeps (all densities in g/cm^3):")

# Isothermal: fuel and moderator at same T
print("\n  Isothermal sweep (fuel = moderator):")
print(f"    T = {T_iso_min:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_iso_min):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_iso_min):.4f}")
print(f"    T = {T_iso_max:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_iso_max):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_iso_max):.4f}")

# Fuel sweep (FTC): fuel varies, moderator fixed
print(f"\n  Fuel sweep (FTC): moderator fixed at {T_mod_fixed:.1f} K")
print(f"    Fuel T = {T_fuel_min:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_fuel_min):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_mod_fixed):.4f}")
print(f"    Fuel T = {T_fuel_max:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_fuel_max):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_mod_fixed):.4f}")

# Moderator sweep (MTC): moderator varies, fuel fixed
print(f"\n  Moderator sweep (MTC): fuel fixed at {T_fuel_fixed:.1f} K")
print(f"    Mod T = {T_mod_min:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_fuel_fixed):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_mod_min):.4f}")
print(f"    Mod T = {T_mod_max:.1f} K:"
      f" rho_fuel = {fuel_density_g_cm3(T_fuel_fixed):.4f},"
      f" rho_mod = {water_density_15MPa_g_cm3(T_mod_max):.4f}")
print()

# ----------------------------------------------------------------------
# Central helper: set temps + densities consistently for each run
# ----------------------------------------------------------------------
def set_thermal_state(T_fuel: float,
                      T_mod: float,
                      T_struct: float,
                      T_default_interp: float) -> None:
    """
    Set temperatures and densities for fuel, moderator, and structural
    materials in a consistent way for a given case.

    Fuel and moderator densities are computed from correlations, not
    hard-coded constants.
    """
    # Fuel
    fuel.temperature = T_fuel
    fuel.set_density('g/cm3', fuel_density_g_cm3(T_fuel))

    # Moderator (water)
    water.temperature = T_mod
    water.set_density('g/cm3', water_density_15MPa_g_cm3(T_mod))

    # Structural / others
    cladding.temperature = T_struct
    ctrl_rod.temperature = T_struct
    rpv.temperature      = T_struct

    # Default temperature for interpolation
    settings.temperature['default'] = T_default_interp

# ----------------------------------------------------------------------
# 1) Isothermal temperature sweep: T_fuel = T_mod = T_clad
# ----------------------------------------------------------------------
keff_means_iso = []
keff_stds_iso  = []
rho_pcm_iso    = []
rho_pcm_err_iso = []

for T in temps_iso:
    # All materials at the same temperature
    set_thermal_state(T_fuel=T,
                      T_mod=T,
                      T_struct=T,
                      T_default_interp=T)

    sp_filename = model.run()
    sp = openmc.StatePoint(sp_filename, autolink=False)

    try:
        k_var  = sp.keff        # newer API: uncertainties.UFloat
        k_mean = k_var.nominal_value
        k_std  = k_var.std_dev
    except AttributeError:
        k_mean, k_std = sp.k_combined  # older API

    sp.close()

    keff_means_iso.append(k_mean)
    keff_stds_iso.append(k_std)

    # Reactivity ρ = (k - 1)/k ; dρ/dk = 1/k^2
    rho     = (k_mean - 1.0) / k_mean
    rho_err = k_std / (k_mean ** 2)

    rho_pcm_iso.append(rho * 1.0e5)
    rho_pcm_err_iso.append(rho_err * 1.0e5)

temps_iso       = np.array(temps_iso)
keff_means_iso  = np.array(keff_means_iso)
keff_stds_iso   = np.array(keff_stds_iso)
rho_pcm_iso     = np.array(rho_pcm_iso)
rho_pcm_err_iso = np.array(rho_pcm_err_iso)

# ----------------------------------------------------------------------
# Plot isothermal reactivity vs temperature
# ----------------------------------------------------------------------
plt.figure()
plt.errorbar(temps_iso, rho_pcm_iso, yerr=rho_pcm_err_iso, fmt="o-")
plt.xlabel("Isothermal temperature T (fuel = moderator = cladding) [K]")
plt.ylabel("Reactivity ρ [pcm]")
plt.title("Isothermal temperature reactivity")
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------
# Fit isothermal temperature coefficient
# ----------------------------------------------------------------------
alpha_T_iso = None
if temps_iso.size >= 2:
    coeffs = np.polyfit(temps_iso, rho_pcm_iso, 1)  # ρ = a T + b
    alpha_T_iso = coeffs[0]                         # pcm/K
    print("Isothermal temperature reactivity coefficient:")
    print(f"  alpha_T_iso ≈ {alpha_T_iso:.3f} pcm/K")
else:
    print("Not enough isothermal points to fit a coefficient.")

# ----------------------------------------------------------------------
# 2) Fuel temperature coefficient (FTC): vary fuel T, fix moderator T
# ----------------------------------------------------------------------
rho_pcm_fuel     = []
rho_pcm_err_fuel = []

for T_fuel in temps_fuel:
    # Fuel varies, moderator & structure fixed at reference
    set_thermal_state(T_fuel=T_fuel,
                      T_mod=T_mod_fixed,
                      T_struct=T_struct_FTC,
                      T_default_interp=T_fuel)

    sp_filename = model.run()
    sp = openmc.StatePoint(sp_filename, autolink=False)

    try:
        k_var  = sp.keff
        k_mean = k_var.nominal_value
        k_std  = k_var.std_dev
    except AttributeError:
        k_mean, k_std = sp.k_combined

    sp.close()

    rho     = (k_mean - 1.0) / k_mean
    rho_err = k_std / (k_mean ** 2)

    rho_pcm_fuel.append(rho * 1.0e5)
    rho_pcm_err_fuel.append(rho_err * 1.0e5)

temps_fuel        = np.array(temps_fuel)
rho_pcm_fuel      = np.array(rho_pcm_fuel)
rho_pcm_err_fuel  = np.array(rho_pcm_err_fuel)

alpha_T_fuel = None
if temps_fuel.size >= 2:
    coeffs_fuel = np.polyfit(temps_fuel, rho_pcm_fuel, 1)
    alpha_T_fuel = coeffs_fuel[0]
    print("\nFuel temperature reactivity coefficient (FTC):")
    print(f"  alpha_T_fuel ≈ {alpha_T_fuel:.3f} pcm/K")
else:
    print("Not enough fuel-temperature points to fit FTC.")

# ----------------------------------------------------------------------
# 3) Moderator temperature coefficient (MTC): vary moderator T, fix fuel T
# ----------------------------------------------------------------------
rho_pcm_mod     = []
rho_pcm_err_mod = []

for T_mod in temps_mod:
    # Moderator varies, fuel & structure fixed
    set_thermal_state(T_fuel=T_fuel_fixed,
                      T_mod=T_mod,
                      T_struct=T_struct_MTC,
                      T_default_interp=T_mod)

    sp_filename = model.run()
    sp = openmc.StatePoint(sp_filename, autolink=False)

    try:
        k_var  = sp.keff
        k_mean = k_var.nominal_value
        k_std  = k_var.std_dev
    except AttributeError:
        k_mean, k_std = sp.k_combined

    sp.close()

    rho     = (k_mean - 1.0) / k_mean
    rho_err = k_std / (k_mean ** 2)

    rho_pcm_mod.append(rho * 1.0e5)
    rho_pcm_err_mod.append(rho_err * 1.0e5)

temps_mod        = np.array(temps_mod)
rho_pcm_mod      = np.array(rho_pcm_mod)
rho_pcm_err_mod  = np.array(rho_pcm_err_mod)

alpha_T_mod = None
if temps_mod.size >= 2:
    coeffs_mod = np.polyfit(temps_mod, rho_pcm_mod, 1)
    alpha_T_mod = coeffs_mod[0]
    print("\nModerator temperature reactivity coefficient (MTC):")
    print(f"  alpha_T_mod ≈ {alpha_T_mod:.3f} pcm/K")
else:
    print("Not enough moderator-temperature points to fit MTC.")

# ----------------------------------------------------------------------
# NEW: plots for α_T_fuel and α_T_mod sweeps (ρ vs T for FTC/MTC)
# ----------------------------------------------------------------------

# Fuel temperature sweep plot
plt.figure()
plt.errorbar(temps_fuel, rho_pcm_fuel, yerr=rho_pcm_err_fuel, fmt="o-")
plt.xlabel("Fuel temperature T_fuel [K]")
plt.ylabel("Reactivity ρ [pcm]")
if alpha_T_fuel is not None:
    plt.title(f"Fuel temperature reactivity (FTC)\nα_T,fuel ≈ {alpha_T_fuel:.2f} pcm/K")
else:
    plt.title("Fuel temperature reactivity (FTC)")
plt.grid(True)
plt.tight_layout()

# Moderator temperature sweep plot
plt.figure()
plt.errorbar(temps_mod, rho_pcm_mod, yerr=rho_pcm_err_mod, fmt="o-")
plt.xlabel("Moderator temperature T_mod [K]")
plt.ylabel("Reactivity ρ [pcm]")
if alpha_T_mod is not None:
    plt.title(f"Moderator temperature reactivity (MTC)\nα_T,mod ≈ {alpha_T_mod:.2f} pcm/K")
else:
    plt.title("Moderator temperature reactivity (MTC)")
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Core geometry (from LOCA model)
# -----------------------------
R_core_cm = 70.0          # cm (LOCA core radius)
H_core_cm = 225.0         # cm (half-height); total height = 2 * H_core_cm

R_core = R_core_cm * 1e-2      # m
H_core_half = H_core_cm * 1e-2 # m
H_core_tot = 2.0 * H_core_half # m

# Approximate core lateral surface area (heat transfer area)
A_core = 2.0 * np.pi * R_core * H_core_tot  # m^2

# ---------------------------------
# Point kinetics + thermal feedback
# ---------------------------------
beta_eff = 0.0065
Lambda   = 1.0e-3       # s, *effective* prompt neutron generation time
                        # (artificially large to reduce stiffness)

# 6-group delayed neutron data (typical PWR)
beta_i   = np.array([0.000215, 0.001424, 0.001274,
                     0.002568, 0.000748, 0.000271])
lambda_i = np.array([0.0124, 0.0305, 0.111, 0.301, 1.14, 3.01])

assert abs(beta_i.sum() - beta_eff) < 1e-4

# Temperature feedback: ρ(T) = ρ_ext(t) + α_T (T_f - T_ref)
T_ref   = 600.0      # K (reference fuel temperature)
alpha_T = -1.6e-5    # Δk/k per K  (≈ -5 pcm/K)

# -----------------------------
# External reactivities
# -----------------------------
# Target initial k_eff at t = 0 (with rods in, before LOCA, T_f = T_ref).
# You can change this value.
k0_target = 1

# Reactivity corresponding to this k_eff:
#   ρ0 = (k0 - 1) / k0  (Δk/k units)
rho0_target = (k0_target - 1.0) / k0_target

# Constant rod worth (control rods in).
# At t=0, LOCA has not yet occurred and T_f = T_ref, so:
#   ρ(0) = ρ_CR = ρ0_target
rho_CR = rho0_target

# LOCA reactivity change (step at t_step)
rho_step = -300.0e-5    # -300 pcm
t_step   = 1.0          # s, time at which LOCA reactivity step is applied

# -----------------------------
# Thermal parameters (lumped fuel)
# -----------------------------
P_nom   = 20.0e6      # W, nominal thermal power
M_fuel  = 4.80e6       # kg, effective fuel mass
cp_fuel = 300.0       # J/kg-K
T_cool  = 600.0       # K, effective coolant/steam temperature in LOCA
h_core  = 100.0       # W/m^2-K, low effective core heat transfer coefficient, VERY APPROXIMATE

# -----------------------------
# Time integration parameters
# -----------------------------
t_end   = 30.0   # s
dt      = 1.0e-3 # s
n_steps = int(t_end / dt)

t     = np.linspace(0.0, t_end, n_steps + 1)
P     = np.zeros_like(t)       # relative power P/P_nom
T_f   = np.zeros_like(t)       # core-average fuel temperature [K]
k_eff = np.zeros_like(t)       # multiplication factor

C = np.zeros((6, n_steps + 1))  # delayed neutron precursor concentrations

# -----------------------------
# Total reactivity function
# -----------------------------
def total_reactivity(t_cur, T_f_cur):
    """
    Total reactivity:
        ρ(t) = ρ_CR + ρ_LOCA(t) + α_T (T_f - T_ref)

    At t < t_step: ρ_LOCA = 0
    At t ≥ t_step: ρ_LOCA = rho_step
    """
    rho_loca = rho_step if t_cur >= t_step else 0.0
    rho_fb   = alpha_T * (T_f_cur - T_ref)
    return rho_CR + rho_loca + rho_fb

# -----------------------------
# Initial conditions
# -----------------------------
P[0]   = 1.0       # start from nominal relative power
T_f[0] = T_ref     # initial fuel temperature

# Precursor concentrations (approximate steady-state for ρ ≈ ρ_CR)
for g in range(6):
    C[g, 0] = beta_i[g] / (Lambda * lambda_i[g]) * P[0]

# Compute and print initial reactivity and k_eff from the chosen rho_CR
rho_init = total_reactivity(0.0, T_f[0])
k_init   = 1.0 / (1.0 - rho_init)
k_eff[0] = k_init

print("Initial state from chosen k0_target:")
print(f"  k0_target (requested) = {k0_target:.5f}")
print(f"  ρ_init (computed)     = {rho_init*1.0e5:.1f} pcm")
print(f"  k_eff(0) (computed)   = {k_init:.5f}\n")

# explicit Euler time integration
for n in range(n_steps):
    rho = total_reactivity(t[n], T_f[n])

    # k_eff from ρ = (k - 1)/k  ⇒  k = 1 / (1 - ρ)
    k_eff[n] = 1.0 / (1.0 - rho)

    # Point kinetics equations
    dPdt = ((rho - beta_eff) / Lambda) * P[n] + np.sum(lambda_i * C[:, n])
    dCdt = beta_i / Lambda * P[n] - lambda_i * C[:, n]

    # Lumped fuel energy balance:
    #   M_fuel * c_p * dT_f/dt = Q_gen - Q_loss
    Q_gen  = P_nom * P[n]
    Q_loss = h_core * A_core * (T_f[n] - T_cool)
    dTdt   = (Q_gen - Q_loss) / (M_fuel * cp_fuel)

    # Explicit Euler step
    P[n + 1]    = max(P[n] + dt * dPdt, 0.0)  # avoid negative power
    C[:, n + 1] = C[:, n] + dt * dCdt
    T_f[n + 1]  = T_f[n] + dt * dTdt

# k_eff at final time
rho_final   = total_reactivity(t[-1], T_f[-1])
k_eff[-1]   = 1.0 / (1.0 - rho_final)
P_rel_final = P[-1]
T_f_final   = T_f[-1]

print("Final state summary:")
print(f"  P_rel(final)        = {P_rel_final:.3f}")
print(f"  T_f(final) [K]      = {T_f_final:.1f}")
print(f"  k_eff(final)        = {k_eff[-1]:.5f}")

# -----------------------------
# Plots: power(t) and T_f(t)
# -----------------------------

# 1) Power vs time
plt.figure()
plt.plot(t, P, label='Relative power P/P_nom')
plt.axvline(t_step, color='k', linestyle='--', linewidth=0.8,
            label='LOCA step')
plt.xlabel('Time [s]')
plt.ylabel('Relative power [-]')
plt.grid(True)
plt.legend()
plt.tight_layout()

# 2) Core-average fuel temperature vs time
plt.figure()
plt.plot(t, T_f, label='Fuel temperature')
plt.axvline(t_step, color='k', linestyle='--', linewidth=0.8,
            label='LOCA step')
plt.xlabel('Time [s]')
plt.ylabel('Core-average fuel temperature [K]')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
